## Lumo ##
Lumo biedt twee opties aan, één via openpyxl, één via pandas

#### Optie 1: met openpyxl (lichtgewicht): ####

pip install openpyxl

```python
from openpyxl import Workbook

def list_to_excel(data_list, filename="LumoOpenpyxl.xlsx"):
    wb = Workbook()
    ws = wb.active
    for i, item in enumerate(data_list, start=1):
        ws.cell(row=i, column=1, value=item)
    wb.save(filename)

mijn_list = ["appel", "peer", "banaan"]
list_to_excel(mijn_list)
```

#### Optie 2: met pandas (handig voor grotere datasets): ####

```python
pip install openpyxl pandas
```

```python
import pandas as pd

def list_to_excel(data_list, filename="LumoPandas.xlsx"):
    df = pd.DataFrame(data_list, columns=["Data"])
    df.to_excel(filename, index=False)

mijn_list = ["appel", "peer", "banaan"]
list_to_excel(mijn_list)
```


## Claude ##
Claude biedt drie opties aan, één via openpyxl, twee via pandas waarvan eentje het wegschrijven naar meerdere tabbladen.

#### Optie 1: met openpyxl ####

```python
from openpyxl import Workbook

data = [10, 20, 30, 40, 50]

wb = Workbook()
ws = wb.create_sheet("Probeersel")
for item in data:
    ws.append([item])
wb.save("ClaudeOpenpyxl.xlsx")
```

#### Optie 2: met pandas, eenvoudig 
Opgelet, deze code overschrijft de volledige spreadsheet.

```python
import pandas as pd

data = [10, 20, 30, 40, 50]

pd.DataFrame(data, columns=["Waarden"]).to_excel("ClaudePandas.xlsx", sheet_name="Probeersel", index=False)
```

#### Optie 3: met pandas, minder eenvoudig 
Meerdere tabs bewaren.
Opgelet, xlsx bestand MOET al bestaan.

```python
import pandas as pd

data = [10, 20, 30, 40, 50]

with pd.ExcelWriter("ClaudePandas.xlsx", engine="openpyxl", mode="a") as writer:
    pd.DataFrame(data, columns=["Waarden2"]).to_excel(writer, sheet_name="Probeersel2", index=False)
```
